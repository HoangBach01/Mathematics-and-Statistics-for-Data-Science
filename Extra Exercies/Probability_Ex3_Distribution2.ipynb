```python
temp = data2['Temp']
stats.describe(temp)
```

```python
temp.mean()
```

```python
temp.median()
```

```python
temp.mode()
```

```python
lst = [0, 1, 2, 2.5, 97.5, 98, 99, 100]
percentile_result = list(map(lambda x: np.percentile(temp, x), lst))
percentile_result
```

```python
temperatures = stats.norm.rvs(loc=temp.mean(), scale=temp.std(), size=10000)
temperatures[:10]
```

```python
sb.distplot(temperatures);
```

```python
sigma = temperatures.std()
three_sigmas = [sigma, sigma * 2, sigma * 3]
three_sigmas
```

```python
z_Hoa = (99.5 - temperatures.mean())/sigma
z_Hoa
```

```python
stats.norm.cdf(z_Hoa)
```
