```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.decomposition import PCA
```

```python
data = pd.read_excel('../Data/B3/Iris.xls')
data
```

```python
corr = data.corr()
corr
```

```python
sb.heatmap(data=corr, cmap='coolwarm');
```

```python
df = data.iloc[:,[0,1,2,3]]
df
```

```python
pca = PCA(2)
pca.fit(df)
```

```python
T = pca.transform(df)
transform_data = pd.DataFrame(T, columns=['1st Component', '2nd Component'])
```

```python
plt.figure(figsize=(12,10))
sb.scatterplot(data=transform_data, x='1st Component', y='2nd Component', hue=data['iris'], s=80);
```

```python
print('Explain percent in total component: '+ str(np.round(np.sum(pca.explained_variance_ratio_)*100,2)) + str('%'))
```
