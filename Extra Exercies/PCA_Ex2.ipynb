```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_excel('../Data/B3/student.xlsx')
df
```

```python
pca = PCA().fit(df)
```

```python
k = pd.DataFrame({'Dimension': np.arange(1,pca.n_components_ + 1),'Cum ratio': np.cumsum(pca.explained_variance_ratio_)})
k
```

```python
sb.lineplot(data=k, x='Dimension', y='Cum ratio');
```

```python
pca = PCA(2)
P = pca.fit_transform(df)
P
```

```python
df2 = pd.DataFrame(P, columns=['Component 1', 'Component 2'])
df2
```

```python
sb.jointplot(data=df2, x='Component 1', y='Component 2');
```
