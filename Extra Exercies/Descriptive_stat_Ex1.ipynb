```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats
```

```python
cars = np.loadtxt('../Data/B5/count_cars.txt').reshape(1,-1)[0]
cars
```

```python
sb.distplot(cars);
```

```python
print('Mean: %.2f , Median: %.2f , \nMode: %.2f w count %.2f, \nMin: %.2f , Max %.2f, \nRange: %.2f' %(np.mean(cars), np.median(cars), stats.mode(cars)[0], stats.mode(cars)[1], cars.min(), cars.max(), cars.max()-cars.min() ))
```

```python
percentile_lst = np.array([5, 15, 25, 50, 75, 90])
percentile_result = np.array(list(map(lambda x: np.percentile(cars, x), percentile_lst)))
percentile_result
```

```python
plt.scatter(x=percentile_result, y=percentile_lst)
plt.xlabel('Percentile Numbers')
plt.ylabel('Percentile Threshold')
plt.show()
```

```python
plt.figure(figsize=(4,6))
sb.boxplot(y=cars)
plt.show()
```

```python
z = stats.zscore(cars)
```

```python
for i in np.where((z>3) | (z<-3))[0]:
    print('Index of Outlier: %i, Outlier Value: %i' %(i, cars[i]) )
```

```python
IQR = stats.iqr(cars)
IQR
```

```python
cars.var()
```

```python
cars.std()
```

```python
stats.skew(cars)
# Nhận xét: Phân phối của cars có độ xiên tương đối gần với phân phối chuẩn (hệ số skewness gần bằng 0), chỉ bị lệch nhẹ sang phải (Mean > Median > Mode)
```

```python
stats.kurtosis(cars)
# Nhận xét: Phân phối của cars nhọn hơn phân phối chuẩn vì hệ số kurtosis lớn hơn 0
```
