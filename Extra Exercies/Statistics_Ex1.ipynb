```python
# Cau 4:
data_trees = stats.norm.rvs(loc=7.5, scale=2.3, size=100)
data_trees[:10]
```

```python
x1, x2 = stats.norm.interval(0.95, loc=data_trees.mean(), scale=stats.sem(data_trees, ddof=0))
```

```python
stats.t.interval(0.95, loc=data_trees.mean(), scale=stats.sem(data_trees, ddof=0), df=(data.shape[0]-1))
```

```python
plt.figure(figsize=(12,6))
sb.distplot(data_trees, bins=10)
plt.axvline(x=x1, linestyle='--', linewidth=3, label='Confidence Interval', color='orange')
plt.axvline(x=x2, linestyle='--', linewidth=3, color='orange')
plt.legend()
plt.show()
```
